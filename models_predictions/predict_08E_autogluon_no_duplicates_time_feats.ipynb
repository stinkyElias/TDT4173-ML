```python
import pandas as pd

train = pd.read_csv(os.path.join(load_data_path, 'no_duplicates_time_feats_train.csv'))
test = pd.read_csv(os.path.join(load_data_path, 'no_duplicates_time_feats_test.csv'))

# Drop features day,day_of_year,month,hour,building in train and day,day_of_year,month,hour in test
# train = train.drop(['day', 'day_of_year', 'month', 'hour', 'building'], axis=1)
# test = test.drop(['day', 'day_of_year', 'month', 'hour'], axis=1)
```

AutoGluon parameters

```python
label = 'pv_measurement'
metric = 'mean_absolute_error'
path = os.path.join(saved_model_path, '08E_no_duplicates_time_feats')
```

```python
predictor = TabularPredictor(label='pv_measurement',
                            #  problem_type='regression',
                             eval_metric=metric,
                             path=path,
                            #  verbosity=3,
                            #  sample_weight=None,
                            #  weight_evaluation=False
                             )

predictor.fit(train_data=train,
            #   presets='best_quality'
              )
```

```python
predictor.leaderboard()
```

```python
predictor.fit_summary(show_plot=True)
```

```python
predictor.feature_importance(data=train)
```

```python
best_model = predictor.get_model_best()
print(best_model)
```

```python
test = TabularDataset(test)

y_pred = predictor.predict(test)
```
