```python
%autoreload 2

import sys
import os

sys.path.append('../')

from autogluon.tabular import TabularDataset, TabularPredictor

home_dir = os.path.join(os.path.expanduser('~'), 'Documents')
saved_data_path = os.path.join(home_dir, 'TDT4173-ML/data_results')
saved_model_path = os.path.join(home_dir, 'models/AutoGluonModels')
load_data_path = os.path.join(home_dir, 'TDT4173-ML/data_feature_engineering')
```

```python
import pandas as pd

train = pd.read_csv(os.path.join(load_data_path, 'no_duplicates_time_feats_train.csv'))
test = pd.read_csv(os.path.join(load_data_path, 'no_duplicates_time_feats_test.csv'))

train = train.drop(['day', 'day_of_year', 'month', 'hour', 'building','fresh_snow_12h:cm', 'fresh_snow_1h:cm', 'fresh_snow_24h:cm',
                    'fresh_snow_3h:cm','fresh_snow_6h:cm', 'snow_density:kgm3', 'snow_depth:cm', 'snow_drift:idx','snow_melt_10min:mm',
                    'snow_water:kgm2'], axis=1)
test = test.drop(['day', 'day_of_year', 'month', 'hour','fresh_snow_12h:cm', 'fresh_snow_1h:cm', 'fresh_snow_24h:cm', 'fresh_snow_3h:cm',
                  'fresh_snow_6h:cm', 'snow_density:kgm3', 'snow_depth:cm', 'snow_drift:idx','snow_melt_10min:mm', 'snow_water:kgm2', 'date_calc'], axis=1)
```

AutoGluon parameters

```python
label = 'pv_measurement'
metric = 'mean_absolute_error'
path = os.path.join(saved_model_path, '09E_no_duplicates_no_snow_best_quality')
```

```python
predictor = TabularPredictor(label='pv_measurement',
                            #  problem_type='regression',
                             eval_metric=metric,
                             path=path,
                             verbosity=3,
                            #  sample_weight=None,
                            #  weight_evaluation=False
                             )

predictor.fit(train_data=train,
              presets='best_quality'
              )
```

```python
predictor.leaderboard()
```

```python
predictor.fit_summary(show_plot=True)
```

```python
predictor.feature_importance(data=train)
```

```python
best_model = predictor.get_model_best()
print(best_model)
```

```python
test = TabularDataset(test)

y_pred = predictor.predict(test)
```

```python
import pandas as pd

predictions = pd.DataFrame(y_pred)

# predictions.head()
# predictions.reset_index(inplace=True)
# predictions.drop(columns=['date_forecast'], inplace=True)

# # Set all negative values to zero
predictions[predictions < 0] = 0

predictions.to_csv(os.path.join(saved_data_path, '09E_autogluon_no_duplicates_no_snow_best_quality.csv'), index=True, index_label='id', header=['predictions'])
```
